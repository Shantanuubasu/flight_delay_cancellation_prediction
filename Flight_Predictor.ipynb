```python
irrelevant_columns = ["FlightNumber", "TailNumber", "AircraftType"]  
df.drop(columns=irrelevant_columns, inplace=True, errors='ignore')
```

```python
encoded_df = df.copy()
categorical_columns = ['FL_DATE','AIRLINE','AIRLINE_DOT','AIRLINE_CODE','ORIGIN','ORIGIN_CITY','DEST','DEST_CITY','CANCELLATION_CODE']
for col in categorical_columns:
    encoded_df[col] = encoded_df[col].astype('category').cat.codes
```

```python
correlation_matrix = encoded_df.corr()
plt.figure(figsize=(20, 12))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()
```
